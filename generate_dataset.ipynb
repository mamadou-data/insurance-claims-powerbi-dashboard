```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python


# Nombre de lignes
n = 8000

np.random.seed(42)

# Listes de valeurs possibles
regions = ["Île-de-France", "Occitanie", "Nouvelle-Aquitaine", "PACA", "Hauts-de-France"]
claim_types = ["Dégât des eaux", "Incendie", "Vol", "Bris de glace", "Catastrophe naturelle"]
policy_types = ["Habitation", "Auto", "Professionnelle"]
experts = [f"Expert_{i}" for i in range(1, 16)]

# Génération dates
start_date = datetime(2023, 1, 1)

data = []

for i in range(n):
    declaration_date = start_date + timedelta(days=np.random.randint(0, 365))
    expertise_delay = np.random.randint(1, 20)
    closure_delay = expertise_delay + np.random.randint(5, 60)
    
    expertise_date = declaration_date + timedelta(days=expertise_delay)
    closure_date = declaration_date + timedelta(days=closure_delay)

    estimated_damage = np.random.gamma(shape=2, scale=1500)
    paid_amount = estimated_damage * np.random.uniform(0.7, 1.0)

    fraud_flag = np.random.choice([0, 1], p=[0.95, 0.05])

    status = np.random.choice(["Clos", "Ouvert", "En recours"], p=[0.7, 0.2, 0.1])

    data.append([
        f"CLM_{i+1}",
        f"CUST_{np.random.randint(1000, 5000)}",
        random.choice(regions),
        random.choice(claim_types),
        random.choice(policy_types),
        declaration_date,
        expertise_date,
        closure_date if status == "Clos" else None,
        status,
        round(estimated_damage, 2),
        round(paid_amount, 2),
        random.choice(experts),
        fraud_flag
    ])

columns = [
    "claim_id",
    "client_id",
    "region",
    "claim_type",
    "policy_type",
    "declaration_date",
    "expertise_date",
    "closure_date",
    "status",
    "estimated_damage",
    "paid_amount",
    "expert_assigned",
    "fraud_flag"
]

df = pd.DataFrame(data, columns=columns)

# Sauvegarde en CSV
df.to_csv("C:/Users/mdied/Downloads/synthetic_claims.csv", index=False)

print("Dataset généré avec succès :", df.shape)
```
